## AI Statistics - Acquired by advanced Machine/Deep Learning & Artificial Intelligence models

* **Expected Goals (xG)** - Statistical model created with the aim of obtaining a more accurate reflection of what really happened during a game (or season) in relation to the quality of the opportunities and shots that each team or player was able to generate. Unlike traditional statistics (total goals scored, total shots, shots on goal, etc.), xG focuses on the quality of shots rather than quantity, and assigns a probability (based on historical data) for each shot made to finish in a goal. Each scoring opportunity will contribute to a points system. For example, a penalty shot has a 76% chance of being scored, that is, 0.76 points that will be added to the other xG values of other shots made by the same player.

* **Expected Assists (xA)** - Statistical model created with the purpose of measuring the danger of a pass made by a certain player that ends in a shot by a teammate, assigning a probability that the created opportunity ends in a goal (also based on historical data). The metric also measures the quality of passes by taking into account the quality of the finisher (the one who takes the shot after receiving the pass).

* **Expected Threat (xT)** - Statistical model that divides the field into a grid, with each cell assigned a probability that an action initiated there will result in a goal in the next N actions. The approach allows us to assess not only the parts of the field from which direct scoring is most likely, but also those from which an assist is most likely to occur. Actions that move the ball, such as passes and dribbles (also called ball runs), can be judged solely based on their start and end points, taking the difference in xT between the start and end cell.

* **Expected Points (xPTS)** - Statistical model that measures the number of points a team could have expected to gain from a game, based on the Goal Opportunities (xG) they created and conceded in that game. It is a cumulative score that can be used as a reference at the end of the season to see which teams deserved perhaps more or less points than they had.

* **The models continue to develop and will be a fundamental part of the future of football, those that adapt to change will have a significant competitive advantage.**

However, filtering based on Expected Assists Peña rises to high positions only below Tannane and Stengs. Without going into technicalities, Peña has generated enough scoring chances to take a total of 5.29 Assists, but had an "under-performance" of -3.29.

What does this mean?

* Peña is a victim of the lack of effectiveness of his teammates.
* If all the passes made by Peña (which ended in scoring opportunities) had been finished off by the average player, Peña would have 5.29 assists. Based on probability and statistics.

##### The importance of understanding these advanced models: 

Unlike an "under-performance" in xG, which is synonymous with a lack of effectiveness in front of goal, a negative performance in xA is interpreted as a lack of effectiveness of the finisher, not the passer. When carrying out a performance analysis, it is necessary to focus on advanced statistical models, since they are metrics that more realistically explain the development of a match or the performance of a player. In this specific case, Peña could be incorrectly classified as a midfielder without the ability to assist in the final meters (based on Total Assists), however the xA model explains that this statement is false, that in fact Peña is one of the best of the league in this facet of the game.

## Multivariable Regression Model (Possession and Dribbling) and xG prediction exercise of a team based on variables.

```python
# load dataset again / importar el set de datos
laliga_df = pd.read_csv("LaLigaDataset_U.csv") 
laliga_df.head()
laliga_df.columns

#define x and y / definir las variables x (variables independientes) e y (variable dependiente / target variable)
x = laliga_df.drop(['League', 'Season', 'Team', 'xG', 'xGA', 'xPTS', 'Press', 'Press_Succ',
       'Press%', 'P_Def_3rd', 'P_Mid_3rd', 'P_Att_3rd', 'Touches',
       'T_Def_3rd', 'T_Mid_3rd', 'T_Att_3rd', 'T_Att_Pen', 'Drib_Succ', 'DribSucc%', '#Players_Dribbled', 'Megs', 'Pass_Cmp',
       'Pass_Att', 'Pass_Cmp%', 'Pass_TotDist', 'Pass_PrgDist'], axis=1).values
y = laliga_df["xG"].values
```

```python
#split the dataset in training and test sets / dividir el dataset en data de entrenamiento y testeo
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)

# train the model / entrenar y configurar el modelo
from sklearn.linear_model import LinearRegression
ml = LinearRegression()
ml.fit(x_train, y_train)

# predict the test results / predecir los resultados del modelo
y_pred = ml.predict(x_test)
print(y_pred)
```

```python
# Evaluate the model / Evaluamos el modelo
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

#create dataframe with test and prediction values / crear un dataframe con la diferencia entre el valor predecido y el actual
pred_df = pd.DataFrame({"Actual Value":y_test, "Predicted Value":y_pred, "Difference":y_test-y_pred})
r2_score(y_test,y_pred)
```

```python
# make a prediction with our input / Predecir en base a nuestro criterio (input)

print("Based on the inputs, the model predicts a final xG score of:", ml.predict([[laliga_df["Poss"].mean(), laliga_df["Drib_Att"].mean()]]))
```

#### Model Interpretation

- The model will make a prediction based on our inputs (Possession Levels and Dribbles made per season). On this occasion we have used the average levels of possession and dribbles made, and basically the model is telling us that having a total possession of 50.043% at the end of the season and about 609 dribbles attempted per season, guarantees you a total xG of 50 goals by season. These models can be very useful in the search for optimization and finally efficiency in terms of results, since we know that the xG is strongly correlated with the total points and especially with the Expected Points, and an increase in these scores is synonymous with being doing things right.

- **Coefficient of determination**: the coefficient of determination in this case is 0.5212, which given the availability of data is an acceptable score.
